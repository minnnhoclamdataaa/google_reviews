```python
column_set = list(set(df['query'].tolist()))
```

```python
column_set
```

```python
dictionary = {}
```

```python
for query in column_set:
    for _, row_2 in df_2.iterrows():
        store_id, subprovince, Title, url = row_2['Store_ID'], row_2['subprovince'], row_2['Title'], row_2['URL']
        print(query)
        if str(query) in str(url):
            dictionary[query] = {'store_id': store_id, 'subprovince': subprovince, 'Title': Title}
            break
```
